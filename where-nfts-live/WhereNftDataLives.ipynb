```python
from datetime import datetime
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
eth_protocols = pd.read_csv('data/eth_protocols.csv', names = ['txs','protocol','contracts','ids'], index_col = 'protocol')
bsc_protocols = pd.read_csv('data/bsc_protocols.csv', names = ['txs','protocol','contracts','ids'], index_col = 'protocol')
```

```python
plot = eth_protocols.plot.pie(subplots=True, figsize=(20, 9), title = 'Ethereum NFTs URL protocols')
```

```python
plot = bsc_protocols.plot.pie(subplots=True, figsize=(20, 9), title = 'Binance Smart Chain NFTs URL protocols')
```

```python
eth_protocol_by_month = pd.read_csv('data/eth_protocol_by_month.csv', names = ['date','txs','protocol','contracts','ids'], index_col = 'date')
bsc_protocol_by_month = pd.read_csv('data/bsc_protocol_by_month.csv', names = ['date','txs','protocol','contracts','ids'], index_col = 'date')
```

```python
eth_protocol_by_month_pivot = eth_protocol_by_month.pivot_table(index=['date'], 
            columns=['protocol'], values='contracts').fillna(0)
eth_protocol_by_month_pivot.plot.area(stacked=True, figsize=(20, 8))
```

```python
eth_protocol_by_month_pivot = eth_protocol_by_month.pivot_table(index=['date'], 
            columns=['protocol'], values='txs').fillna(0)
eth_protocol_by_month_pivot.plot.area(stacked=True, figsize=(20, 8))
```

```python
bsc_protocol_by_month_pivot = bsc_protocol_by_month.pivot_table(index=['date'], 
            columns=['protocol'], values='contracts').fillna(0)
bsc_protocol_by_month_pivot.plot.area(stacked=True, figsize=(20, 8))
```

```python
bsc_protocol_by_month_pivot = bsc_protocol_by_month.pivot_table(index=['date'], 
            columns=['protocol'], values='txs').fillna(0)
bsc_protocol_by_month_pivot.plot.area(stacked=True, figsize=(20, 8))
```

```python
eth_domains = pd.read_csv('data/eth_domains.csv', names = ['txs','protocol','domain','contracts','ids'], index_col = 'domain')
bsc_domains = pd.read_csv('data/bsc_domains.csv', names = ['txs','protocol','domain','contracts','ids'], index_col = 'domain')
```

```python
eth_domains.head(10).plot.pie(y = 'ids', figsize=(10, 10))
```

```python
bsc_domains.head(10).plot.pie(y = 'ids', figsize=(10, 10))
```

```python
plt.figure(figsize=(15,15))
wc = WordCloud(background_color="white", width=700, height=400)
wc.generate_from_frequencies(eth_domains.to_dict()['ids'])

plt.imshow(wc)
plt.axis('off')
```

```python
plt.figure(figsize=(15,15))
wc = WordCloud(background_color="white", width=700, height=400)
wc.generate_from_frequencies(bsc_domains.to_dict()['ids'])

plt.imshow(wc)
plt.axis('off')
```
